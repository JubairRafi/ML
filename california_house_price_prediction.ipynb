```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('housing.csv')
df.head()
```

```python
df_shuffled = df.sample(n=len(df),random_state=1)
```

```python
df_shuffled.head()
```

```python
pd.get_dummies(df_shuffled['ocean_proximity'])
```

```python
df_shuffled.head()
```

```python
df_shuffled.drop('ocean_proximity',axis=1)
```

```python
df_final = pd.concat([df_shuffled.drop('ocean_proximity',axis=1),pd.get_dummies(df_shuffled['ocean_proximity'])],axis=1)
```

```python
df_final
```

```python
df_final.columns
```

```python
df_final = df_final[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income', '<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY',
       'NEAR OCEAN',
       'median_house_value']]
```

```python
df_final
```

```python
df_final.isnull().sum()
```

```python
df_final = df_final.dropna()
len(df_final)
```

```python
df_final.isnull().sum()
```

```python
train , valid, test = np.split(df_final.sample(frac=1),[int(0.6*len(df_final)),int(0.8*len(df_final))])
```

```python
len(train),len(valid),len(test)
```

```python
x_train,y_train = train.to_numpy()[:,:-1],train.to_numpy()[:,-1]
x_val,y_val = valid.to_numpy()[:,:-1],valid.to_numpy()[:,-1]
x_test,y_test = test.to_numpy()[:,:-1],test.to_numpy()[:,-1]
```

```python
x_train,y_train
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler().fit(x_train[:, :8])
```

```python
def preprocessor(X):
  A = np.copy(X)
  A[:, :8] = scaler.transform(A[:, :8])
  return A
```

```python
x_train, x_val, x_test = preprocessor(x_train), preprocessor(x_val), preprocessor(x_test)
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression().fit(x_train,y_train)
y_train_pred = lm.predict(x_train)
y_val_pred = lm.predict(x_val)
```

```python
mse(y_train_pred,y_train,squared = False),mse(y_val_pred,y_val,squared = False)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors=11).fit(x_train,y_train)
y_train_pred_knn = knn.predict(x_train)
y_val_pred_knn = knn.predict(x_val)
```

```python
mse(y_train_pred_knn,y_train,squared = False),mse(y_val_pred_knn,y_val,squared = False)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(max_depth=8).fit(x_train,y_train)
y_train_pred_rfr = rfr.predict(x_train)
y_val_pred_rfr = rfr.predict(x_val)
```

```python
mse(y_train_pred_rfr,y_train,squared = False),mse(y_val_pred_rfr,y_val,squared = False)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(n_estimators = 290).fit(x_train, y_train)
y_train_pred_gbr = gbr.predict(x_train)
y_val_pred_gbr = gbr.predict(x_val)
```

```python
mse(y_train_pred_gbr,y_train,squared = False),mse(y_val_pred_gbr,y_val,squared = False)
```
